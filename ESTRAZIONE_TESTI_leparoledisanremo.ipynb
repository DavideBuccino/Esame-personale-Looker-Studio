```python
import os
import requests
from bs4 import BeautifulSoup
import re

# Step 1: URL delle pagine
base_urls = [
    "https://www.leparoledisanremo.it/anno/2023/",
    "https://www.leparoledisanremo.it/anno/2023/page/2/",
    "https://www.leparoledisanremo.it/anno/2023/page/3/"
]

# Step 2: Cartella per i singoli file
folder_name = "Testi_Sanremo_2023"
os.makedirs(folder_name, exist_ok=True)

# Step 3: Funzione per nomi file validi
def clean_filename(name):
    return re.sub(r'[\\/*?:"<>|]', "", name)

# Step 4: Recupera i link ai testi
song_links = []

print("📥 Recupero link dei testi...")
for url in base_urls:
    res = requests.get(url)
    soup = BeautifulSoup(res.content, "html.parser")
    articles = soup.find_all("article")
    for article in articles:
        a_tag = article.find("a")
        if a_tag and a_tag["href"]:
            song_links.append(a_tag["href"])

print(f"✅ Trovati {len(song_links)} testi da scaricare.\n")

# Step 5: Scarica e salva testi
all_lyrics = []

for i, link in enumerate(song_links, 1):
    res = requests.get(link)
    soup = BeautifulSoup(res.content, "html.parser")

    title_tag = soup.find("h1", class_="entry-title")
    lyrics_tag = soup.find("div", class_="entry-content")

    if title_tag and lyrics_tag:
        title = title_tag.get_text().strip()
        lyrics = lyrics_tag.get_text(separator="\n").strip()

        # Singolo file
        filename = clean_filename(title) + ".txt"
        filepath = os.path.join(folder_name, filename)
        with open(filepath, "w", encoding="utf-8") as f:
            f.write(f"{title}\n\n{lyrics}")

        # Aggiunta al file unico
        all_lyrics.append(f"{title}\n\n{lyrics}\n\n{'='*60}\n")

        print(f"✔️ [{i}/{len(song_links)}] Salvato: {title}")

# Step 6: Salva file unico
with open("sanremo2023_tutti_i_testi.txt", "w", encoding="utf-8") as f:
    f.writelines(all_lyrics)

print("\n🎉 Download completato! Tutti i testi salvati.")
```
