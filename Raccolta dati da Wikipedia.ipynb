QUI CERCO DI OTTENERE LE TABELL, LE IMMAGINI E I GRAFICI DALLA PAGINA WEB DI WIKIPEDIA SANREMO

```python
import os
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL della pagina Wikipedia
url = "https://it.wikipedia.org/wiki/Festival_di_Sanremo#Ascolti_2"

# Creazione della cartella per le immagini
cartella_immagini = "immagini_sanremo"
os.makedirs(cartella_immagini, exist_ok=True)

# Headers per la richiesta HTTP
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

# Richiesta della pagina
response = requests.get(url, headers=headers)
response.raise_for_status()

# Parsing del contenuto HTML
soup = BeautifulSoup(response.content, "lxml")

# Estrazione delle tabelle
tabelle = pd.read_html(response.text)

# Salvataggio delle tabelle in un file Excel
nome_file_excel = "ascolti_sanremo_tabelle.xlsx"
with pd.ExcelWriter(nome_file_excel, engine='openpyxl') as writer:
    for i, tabella in enumerate(tabelle):
        nome_foglio = f"Tabella_{i+1}"
        # Verifica e rimozione MultiIndex se presente
        if isinstance(tabella.columns, pd.MultiIndex):
            tabella.columns = [' '.join(map(str, col)).strip() for col in tabella.columns.values]
        tabella.to_excel(writer, sheet_name=nome_foglio, index=False)

print(f"‚úÖ Salvate {len(tabelle)} tabelle in '{nome_file_excel}'")

# Estrazione e download delle immagini
immagini = soup.find_all("img")
print(f"üîç Trovate {len(immagini)} immagini nella pagina.")

for i, img in enumerate(immagini):
    src = img.get("src")
    if not src:
        continue
    if src.startswith("//"):
        src = "https:" + src
    elif src.startswith("/"):
        src = "https://it.wikipedia.org" + src
    elif not src.startswith("http"):
        continue

    nome_file = os.path.basename(src).split("?")[0]
    nome_file = re.sub(r"[^\w\-_\.]", "_", nome_file)
    percorso_file = os.path.join(cartella_immagini, nome_file)

    try:
        img_data = requests.get(src, headers=headers).content
        with open(percorso_file, "wb") as f:
            f.write(img_data)
        print(f"‚úÖ Immagine salvata: {nome_file}")
    except Exception as e:
        print(f"‚ùå Errore nel salvare l'immagine {nome_file}: {e}")
```
